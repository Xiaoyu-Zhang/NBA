```python
def Search_Player_ID(FirstName, LastName, Position = None):
    if not Position:
        query = {'api_key':api_key, 'first_name':FirstName, 'last_name':LastName}
    else:
        query = {'api_key':api_key, 'first_name':FirstName, 'last_name':LastName, 'position':Position}
    tmp = requests.post(url_NBA_player, data = query)
    tmp_dict = eval(tmp.content[1:-1])
    return tmp_dict['player_id']
```

```python
def Search_Player_To_Play(contest_id):
    query = {'api_key':api_key, 'contest_id':contest_id}
    tmp = requests.post(url_draftkings_player, data = query)
    tmp_dict = eval(tmp.content[1:-1])
    t = pd.DataFrame.from_records(eval(tmp.content[1:-1]))
    return t['player_id']
```

```python
def Draftkings_Rule(boxdata):
    score = 0
    score += boxdata['pts'].iloc[0]
    score += 0.5 * boxdata['fg3m'].iloc[0]
    score += 1.25 * (boxdata['dreb'].iloc[0] + boxdata['oreb'].iloc[0])
    score += 1.5 * boxdata['ast'].iloc[0]
    score += 2 * int(boxdata['stl'].iloc[0])
    score += 2 * boxdata['blk'].iloc[0]
    score -= 0.5 * boxdata['to'].iloc[0]
    tmp = sum([boxdata['pts'].iloc[0] >= 10, (boxdata['dreb'].iloc[0] + boxdata['oreb'].iloc[0]) >= 10, boxdata['ast'].iloc[0] >= 10, 
               int(boxdata['stl'].iloc[0]) >= 10, boxdata['blk'].iloc[0] >= 10])
    if tmp >= 3:
        score += 3 + 1.5
    elif tmp == 2:
        score += 1.5
    return score
```

```python
class Player:
    def __init__(self, player_id):
        self.Player_ID = player_id
        self.Draftkings_ID = self.get_Draftkings_ID()
        self.First_Name = self.get_First_Name()
        self.Last_Name = self.get_Last_Name()
        self.Salary = self.get_Salary()
        self.PPG = self.get_PPG()
        self.Position = self.get_Position()
        self.Team = self.get_Team()
        self.last_N_Game = self.get_last_N_Game(5)
    def get_Draftkings_ID(self):
        pass
    def get_First_Name(self):
        pass
    def get_Last_Name(self):
        pass
    def get_Salary(self):
        pass
    def get_PPG(self):
        pass
    def get_Position(self):
        pass
    def get_Team(self):
        pass
    def get_last_N_Game(self, N):
        query_stats = {'api_key':api_key,'player_id':self.Player_ID}
        tmp = requests.post(url_boxscore_player, data = query_stats)
        t = pd.DataFrame.from_records(eval(tmp.content[1:-1]))
        #print(t['season'])
        res = []
        if N < 0:
            t_tmp = t.loc[t['season'] == '2016']
            #print(t_tmp)
            for i in range(len(t_tmp)):
                res.append(Draftkings_Rule(t_tmp[i:i+1]))
            res.reverse()
            return res

        for i in range(1,N+1):
            #print(t.tail(i)[['pts','dreb','oreb','ast','stl','blk','to','fg3m']])
            res.append(Draftkings_Rule(t.tail(i)))
        
        return res
    
```
